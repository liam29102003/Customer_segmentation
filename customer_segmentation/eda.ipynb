```python
features = ["Age", "Income", "Total_Spending", "NumWebPurchases", "NumStorePurchases", "NumWebVisitsMonth","Recency"]
```

```python
X = df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler =  StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i  in range(2,10):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,10), wcss, marker = "o")
plt.title("Elbow method for optimal K")
plt.xlabel("Number of CLusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
df['cluster']
```

```python
cluster_summary = df.groupby("cluster")[features].mean()
```

```python
cluster_summary
```

```python
df['cluster'].value_counts()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
pca_data = pca.fit_transform(X_scaled)
df['PCA1'], df['PCA2'] = pca_data[:,0], pca_data[:,1]
```

```python
pca_data
```

```python
df['PCA1']
```

```python
sns.scatterplot(x="PCA1", y = "PCA2", hue="cluster", data = df, palette = "Set1")
plt.title("customer segmentation (PCA) ")
plt.show()
```

```python
cluster_labels = {
    0: "Low-Spending Seniors",
    1: "Bargain-Hunting Young Adults",
    2: "Active Mid-Age Shoppers",
    3: "Big Spender Premium Customers",
    4: "Loyal High-Spending Customers",
    5: "Occasional Low-Spending Customers"
}
```

```python
cluster_summary['Cluster_Label'] = cluster_summary.index.map(cluster_labels)
```
